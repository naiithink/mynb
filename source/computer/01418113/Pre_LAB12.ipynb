```c++
// ...

class Shape3D
{
public:
    virtual double volumn() {};
    virtual double surfaceArea() {};
};

class RightRectangularPyramid : public Shape3D
{
    public:
        explicit RightRectangularPyramid(const double width = 0, const double length = 0, const double height = 0)
        {
            this->width = width;
            this->length = length;
            this->height = height;
        }

        double volumn()
        {
            double res = 0.0;
            res = (width * length * height) / 3.0;
            return res;
        }

        double surfaceArea()
        {
            double res = 0.0;
            res = ((length * width)
                   + length * sqrt(pow((width / 2.0), 2.0)
                                   + pow(height, 2.0)
                              )
                   + width * sqrt(pow((length / 2.0), 2.0)
                                  + pow(height, 2.0)
                             )
            );
            return res;
        }

    private:
        double width;
        double length;
        double height;
};

class Sphere : public Shape3D
{
    public:
        explicit Sphere(const double radius = 0)
        {
            this->radius = radius;
        }

        double volumn()
        {
            double res = 0;
            res = (4.0 / 3.0) * M_PI * pow(radius, 3.0);
            return res;
        }

        double surfaceArea()
        {
            double res = 0;
            res = 4 * M_PI * pow(radius, 2.0);
            return res;
        }
    
    private:
        double radius;
};

// ...
```
