กำหนดให้

```c
int a=5, b=2;

1. a++%b>>2        /* Ans1 = ? */
2. a|b+3           /* Ans2 = ? */
3. (b<<1)*a        /* Ans3 = ? */
4. a+(7&b)         /* Ans4 = ? */
5. --a*b%3<<2      /* Ans5 = ? */
6. (a+5*b>>1)<<2   /* Ans6 = ? */
```

**หมายเหตุ**<br>
แต่ละคำสั่งไม่มีความต่อเนื่องกัน

````c
##include <stdio.h>

int A, B, C, D, E, F;

int main(void)
{
    int a = 5, b = 2;

    A = a++ % b >> 2                /* Ans1 = 0     */ ;
    //     0011 >> 2

    B = a | b + 3                   /* Ans2 = 5     */ ;
    //       0101
    //    |  0101
    //       0101

    C = (b << 1) * a                /* Ans3 = 20    */ ;
    //      0100
    //         4 * 5

    D = a + (7 & b)                 /* Ans4 = 7     */ ;
    //      0111
    //   &  0010
    //      0010
    //  5 + 2

    E = --a * b % 3 << 2             /* Ans5 = 8    */ ;
    //    4 * 2
    //        8 % 3
    //            2 << 2
    //         0010
    //     <<     2
    //         1000

    F = (a + 5 * b >> 1) << 2       /* Ans6 = 28    */ ;
    //   5 + 10    >> 1
    //          15 >> 1
    //        1111
    //    >>     1
    //        0111       << 2
    //        0111
    //    <<     2
    //   0001 1100

    return 0;
}
````

## 05 ผ่อนคลาย address & dereferencing

เมื่อสิ้นสุดการทำงานของส่วนของโปรแกรมต่อไปนี้

```
int i, j=12;
int *pi, *pj=&j;
*pj = j+3;
i = *pj+7;
pi = pj;
*pi = i+j;
```

สมมติให้ตำแหน่งหน่วยความจำที่เก็บตัวแปร i คือ FF04 และตัวแปร j คือ FF00 จงเติมค่าของตัวแปรหรือนิพจน์ต่อไปนี้

```
&i = 

&j = 

pj = 

*pj = 

i = 

pi = 

*pi = 

pi+2 = 

*pi+2 = 

*pi+*pj = 
```

````c
##include <stdio.h>

int main(void)
{
    int i, j = 12;
    int *pi, *pj = &j;
    
    *pj     =     j + 3    ;        // j  = 15  ::=  j += 3,      pj = &j
      i     =   *pj + 7    ;        // i  = 22  ::=  j += 7
     pi     =    pj        ;        // pi = &j  ::=  pi = pj
    *pi     =     i + j    ;        // j  = 37  ::=  j = i + j
    
    //  i = 22
    //  j = 37
    //  pi = pj = &j = FF00
    //  &i = FF04

    return 0;
}
````
